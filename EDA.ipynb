Merging attributes from all files keeping order_reviews as the master

```python
df_m1 = pd.merge(order_reviews, orders, how='left', on = 'order_id')
df_m2 = pd.merge(df_m1, order_payments, how='left', on = 'order_id')
df_m3 = pd.merge(df_m2, order_items, how='left', on = 'order_id')
df_m4 = pd.merge(df_m3, products, how='left', on = 'product_id')
df_m5 = pd.merge(df_m4, category_name, how='left', on = 'product_category_name')
df_m6 = pd.merge(df_m5, sellers, how='left', on = 'seller_id')
df_final = pd.merge(df_m6, customer_df, how='left', on = 'customer_id')

df_final.head(5)
```

Converting date specific columns to date datatype and deriving features like 'days_diff_est_del' (difference in days between order_estimated_delivery_date and order_delivered_customer_date) and 'days_diff_shp_del' (difference in days between shipping_limit_date and order_delivered_customer_date)

```python
#orders.dtypes
df_mod = df.copy()
df_mod["order_purchase_timestamp"] = pd.to_datetime(df["order_purchase_timestamp"], format='%Y-%m-%d %H:%M:%S')
df_mod["order_delivered_carrier_date"] = pd.to_datetime(df["order_delivered_carrier_date"], format='%Y-%m-%d %H:%M:%S')
df_mod["order_delivered_customer_date"] = pd.to_datetime(df["order_delivered_customer_date"], format='%Y-%m-%d %H:%M:%S')
df_mod["order_estimated_delivery_date"] = pd.to_datetime(df["order_estimated_delivery_date"], format='%Y-%m-%d %H:%M:%S')
df_mod["shipping_limit_date"] = pd.to_datetime(df["shipping_limit_date"], format='%Y-%m-%d %H:%M:%S')
df_mod['days_diff_est_del'] = (df_mod.order_estimated_delivery_date - df_mod.order_delivered_customer_date)/np.timedelta64(1, 'D')
df_mod['days_diff_shp_del'] = (df_mod.shipping_limit_date - df_mod.order_delivered_customer_date)/np.timedelta64(1, 'D')

df_mod[['days_diff_est_del','days_diff_shp_del','shipping_limit_date','order_estimated_delivery_date','order_delivered_customer_date','class']]
```

Deriving features like 'delay_flag' (if 'days_diff_shp_del' is less than 0 days then set 0 else 1) and 'delay_est_flag' (if 'days_diff_est_del' is less than 0 days then set 0 else 1)

---



```python
import numpy as np
df_mod['delay_flag'] = np.where(df_mod['days_diff_shp_del'] <= 0, 0,1)
df_mod['delay_est_flag'] = np.where(df_mod['days_diff_est_del'] <= 0, 0,1)
df_mod
```

Below plot shows that even if there is delay to deliver still those orders have positive reviews so it may indicate that delay is not a key factor for negative review

```python
from matplotlib import pyplot as plt
import seaborn as sns

delay_df_1_tmp = df_mod.loc[df_mod['class'] == 1]
delay_df_0_tmp = df_mod.loc[df_mod['class'] == 0]
delay_df_1 = delay_df_1_tmp['delay_flag'].value_counts().reset_index()
delay_df_0 = delay_df_0_tmp['delay_flag'].value_counts().reset_index()
plt.figure(figsize = (25,7))

plt.subplot(121)
base_color = sns.color_palette()[2]

sns.barplot(data = delay_df_1, x = 'index', y = 'delay_flag', color = base_color)
plt.title('Delay with Positive Reviews')
plt.xlabel('Delay')
plt.ylabel('Status Count')

plt.subplot(122)
base_color = sns.color_palette()[3]
sns.barplot(data = delay_df_0, x = 'index', y = 'delay_flag', color = base_color)
plt.title('Delay with Negative Reviews')
plt.xlabel('Delay')
plt.ylabel('Status Count')
```
